### Bar chart

#### Simple

```python
bar_chart_data.sort_values('Cars Listings',ascending=True).plot(kind='barh',
                                                                 x='Brand',
                                                                 grid=False,
                                                                 figsize=(10,6),
                                                                 title='Cars Listings',
                                                                 xlabel='Car listing',
                                                                 ylabel='Brands',
                                                                 color=palette[0],
                                                                 )
plt.show()
```

#### Pyplot style

```python
plt.figure(figsize = (10, 6))
plt.barh('Brand', 'Cars Listings',
         data = bar_chart_data.sort_values('Cars Listings',ascending=True),
         label='Car Listing',
         color=palette[0])
plt.legend()
plt.xlabel('Car listing')
plt.ylabel('Brands')
plt.title("Cars Listings")
plt.grid(axis='x', linestyle='--', alpha = 0.5)

plt.show()
```

#### Object-oriented style

```python
fig,axes = plt.subplots(nrows=1, ncols=1)

axes.barh(y = 'Brand',
          width = 'Cars Listings',
          data = bar_chart_data.sort_values('Cars Listings',ascending=True),
          label='Car Listing',
          color=palette[0])
axes.set_xlabel('Car listing', fontsize=10)
axes.set_ylabel('Brand', fontsize=10)
axes.set_title('Cars Listings', fontsize=15)
axes.grid(axis='x', linestyle='--', alpha = 0.5)
axes.legend()
fig.set_size_inches(10,6)

plt.show()
```

### Line chart

#### Simple

```python
line_plot_data[(line_plot_data['year']==2019)&(line_plot_data['month']==1)].plot(x='Date', y=['GSPCRet','FTSERet'],
                  kind='line',
                  grid=True,
                  figsize=(10,6),
                  title='GSPCRet and FTSERet line chart for 2019',
                  xlabel='Date', ylabel='Returns',
                  color={"GSPCRet": palette[2], "FTSERet": palette[1]})
plt.show()
```

#### Pyplot style

```python
plt.figure(figsize = (10, 6))
plt.plot('Date', 'GSPCRet', data = line_plot_data[(line_plot_data['year']==2019)&(line_plot_data['month']==1)], color=palette[2])
plt.plot('Date', 'FTSERet', data = line_plot_data[(line_plot_data['year']==2019)&(line_plot_data['month']==1)], color=palette[1])
plt.legend()
plt.xlabel('Date')
plt.ylabel('Returns')
plt.title("GSPCRet and FTSERet line chart for 2019")
plt.xticks(rotation = 45)
plt.grid(linestyle='--', alpha = 0.5)

plt.show()
```

#### Object-oriented style

```python
fig,axes = plt.subplots(nrows=1, ncols=1)

axes.plot('Date', 'GSPCRet', data = line_plot_data[(line_plot_data['year']==2019)&(line_plot_data['month']==1)], color=palette[2])
axes.plot('Date', 'FTSERet', data = line_plot_data[(line_plot_data['year']==2019)&(line_plot_data['month']==1)], color=palette[1])
axes.set_xlabel('Date', fontsize=10)
axes.set_ylabel('Returns', fontsize=10)
axes.set_title('GSPCRet and FTSERet line chart for 2019', fontsize=15)
axes.grid(linestyle='--', alpha = 0.5)
axes.tick_params(axis='x', labelrotation=45)
axes.legend()
fig.set_size_inches(10,6)

plt.show()
```

### Stacked area chart

### Conclusion

- What does the chart show?<br>
The chart illustrates the sum of monthly revenue in 2020 categorized by the source.

- Which would you say is the most important revenue source for the business and why?<br>
The international business is the most important. Not only is it the largest in terms of revenue, but also fluctuations of this segment affect the overall revenue.

- How does the performance of the company look when considering the Date factor?<br>
There are peaks in revenue in months 2, 5, 11, while months 4 and 8 show the lowest revenue.

#### Pyplot style

```python
plt.figure(figsize = (10, 6))
plt.stackplot('year_month', 'International', 'Domestic','Services',
              data = stacked_chart_data,
              colors = [palette[3],palette[1],palette[2]])
plt.legend(labels = ['International', 'Domestic','Services'])
plt.xlabel('Month')
plt.ylabel('Revenue')
plt.title("Revenue by source")
plt.xticks(stacked_chart_data['year_month'], rotation = 45)
plt.ylim(0,10000)
plt.grid(axis='x',linestyle='--', alpha = 0.3)

plt.show()
```

#### Object-oriented style

```python
fig,axes = plt.subplots(nrows=1, ncols=1)

axes.stackplot('year_month', 'International', 'Domestic','Services',
              data = stacked_chart_data,
              colors = [palette[3],palette[1],palette[2]])
axes.set_xlabel('Month', fontsize=10)
axes.set_ylabel('Revenue', fontsize=10)
axes.set_title('Revenue by source', fontsize=15)
axes.grid(axis='x',linestyle='--', alpha = 0.5)
axes.set_ylim(0,10000)
axes.tick_params(axis='x', labelrotation=45)
axes.legend(labels = ['International', 'Domestic','Services'])
fig.set_size_inches(10,6)

plt.show()
```

### Bar and Line chart

#### Object-oriented style

```python
fig,axes = plt.subplots(nrows=1, ncols=1)

axes.bar(x = 'Complaint reasons',
          height = 'Number of complaints',
          data = bar_line_data.sort_values('Number of complaints',ascending=False),
          label='Number of complaints ',
          color=palette[3])

axes.set_xlabel('Complaint reasons', fontsize=10)
axes.set_ylabel('Number of complaints', fontsize=10)
axes.set_title('Number of complaints and its frequency grouped by the reason', fontsize=15)
axes.grid(axis='y', linestyle='--', alpha = 0.5)
axes.legend(loc='best', bbox_to_anchor=(1.0, 1.0))
axes.set_ylim(0,5000)

axes2 = axes.twinx()
axes2.yaxis.set_major_formatter(PercentFormatter(xmax = 1.0))
axes2.plot('Complaint reasons', 'frequency',
           data = bar_line_data.sort_values('Number of complaints',ascending=False),
           label='Frequency', color=palette[1], marker = "o")
axes2.set_ylabel('Frequency', fontsize=10)
axes2.legend(loc='best', bbox_to_anchor=(1.0, 0.9))
axes2.set_ylim(0.,1.3)


fig.set_size_inches(10,6)

plt.show()
```
